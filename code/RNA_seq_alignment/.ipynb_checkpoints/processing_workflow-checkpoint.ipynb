## Merge IFIT3- YTH-TRIBE outputs for common sites

- The bullseye pipeline outputs .bed files with info about each tribe site with the following format:

the first 6 columns being the standard BED6 format: chr start end gene score strand file also contains additional columns: Control_edit_frequency: editing rate in control sample control_covorage: coverage in control samples dart_edit_frequency: editing rate in DART sample (same as score column) dart_coverage: coverage in control samples conversion_type: C2U for DART-seq

since tribe sites are only one nt and will be exact matches, a basic overlap can be done instead of using bedtools

```python
module load Anaconda3/2021.05

source activate ~/hornerlab/mt354/miniconda3/envs/jupyter_python_env
```

```python
import os
import pandas as pd

os.getcwd()
```

```python
os.chdir('MGT450_host_bullseye/matrix_files/')
```

```python
yth_df = pd.read_csv('yth.bed', sep='\t', header=None, usecols=[0,1,2,3,4,5])
header = ['chrom', 'chromStart', 'chromEnd', 'features', 'edit_frequency', 'strand']
yth_df.columns = header[:len(yth_df.columns)]

ifit3_df = pd.read_csv('ifit3.bed', sep='\t', header=None, usecols=[0,1,2,3,4,5])
header = ['chrom', 'chromStart', 'chromEnd', 'features', 'edit_frequency', 'strand']
ifit3_df.columns = header[:len(ifit3_df.columns)]

yth_df.head(10)
```

```python
merged_df = yth_df.merge(ifit3_df, how = 'outer', on = ['chrom', 'chromStart', 'chromEnd','strand'])

merged_df[['gene_x','features_x','other_x']] = merged_df['features_x'].str.split('|',2, expand=True)

merged_df[['gene_y','features_y','other_y']] = merged_df['features_y'].str.split('|',2, expand=True)


merged_df.loc[merged_df['gene_x'] == merged_df['gene_y'], 'gene'] = merged_df['gene_x']
merged_df.loc[merged_df['gene_x'].isnull() == True, 'gene'] = merged_df['gene_y']
merged_df.loc[merged_df['gene_y'].isnull() == True, 'gene'] = merged_df['gene_x']

merged_df.loc[merged_df['features_x'] == merged_df['features_y'], 'feature'] = merged_df['features_x']
merged_df.loc[merged_df['features_x'].isnull() == True, 'feature'] = merged_df['features_y']
merged_df.loc[merged_df['features_y'].isnull() == True, 'feature'] = merged_df['features_x']

merged_df.rename(columns={'edit_frequency_x':'yth_editing','edit_frequency_y':'ifit3_editing'}, inplace=True)

merged_df = merged_df[['gene','feature','yth_editing','ifit3_editing','chrom','chromStart','chromEnd','strand']]

merged_df.to_excel('ifit3_yth_tribe_hcv_common.xlsx', na_rep='NaN')
```

```python
yth_df = pd.read_csv('yth_n9.bed', sep='\t', header=None, usecols=[0,1,2,3,4,5])
header = ['chrom', 'chromStart', 'chromEnd', 'features', 'edit_frequency', 'strand']
yth_df.columns = header[:len(yth_df.columns)]

ifit3_df = pd.read_csv('ifit3_n9.bed', sep='\t', header=None, usecols=[0,1,2,3,4,5])
header = ['chrom', 'chromStart', 'chromEnd', 'features', 'edit_frequency', 'strand']
ifit3_df.columns = header[:len(ifit3_df.columns)]

yth_df.head(10)
```

```python
merged_df = yth_df.merge(ifit3_df, how = 'outer', on = ['chrom', 'chromStart', 'chromEnd','strand'])

merged_df[['gene_x','features_x','other_x']] = merged_df['features_x'].str.split('|',2, expand=True)

merged_df[['gene_y','features_y','other_y']] = merged_df['features_y'].str.split('|',2, expand=True)


merged_df.loc[merged_df['gene_x'] == merged_df['gene_y'], 'gene'] = merged_df['gene_x']
merged_df.loc[merged_df['gene_x'].isnull() == True, 'gene'] = merged_df['gene_y']
merged_df.loc[merged_df['gene_y'].isnull() == True, 'gene'] = merged_df['gene_x']

merged_df.loc[merged_df['features_x'] == merged_df['features_y'], 'feature'] = merged_df['features_x']
merged_df.loc[merged_df['features_x'].isnull() == True, 'feature'] = merged_df['features_y']
merged_df.loc[merged_df['features_y'].isnull() == True, 'feature'] = merged_df['features_x']

merged_df.rename(columns={'edit_frequency_x':'yth_editing','edit_frequency_y':'ifit3_editing'}, inplace=True)

merged_df = merged_df[['gene','feature','yth_editing','ifit3_editing','chrom','chromStart','chromEnd','strand']]

merged_df.to_excel('ifit3_yth_tribe_hcv_common_n9.xlsx', na_rep='NaN')
```
